```python
import torch
import torch.nn as nn
import torchaudio
import math
```

```python
# q1
print("q1. Flatten (32,1,64,64) except batch -> input features?")
x = torch.randn(32, 1, 64, 64)
print("Answer:", x.view(32, -1).shape[1])

# q2
print("\nq2. 1 sec audio (16kHz) -> MelSpec (n_fft=400, hop=160, n_mels=64) shape?")
sr = 16000
wave = torch.randn(1, sr)
mel = torchaudio.transforms.MelSpectrogram(
    sample_rate=sr,
    n_fft=400,
    hop_length=160,
    n_mels=64
)(wave)
print("Answer:", tuple(mel.shape))
```

```python
# q3
print("\nq3. Conv(k=3,s=1,p=1) + MaxPool(2,2) on (64,64) -> spatial size?")
inp = torch.randn(1, 1, 64, 64)
conv = nn.Conv2d(1, 1, 3, 1, 1)
pool = nn.MaxPool2d(2, 2)
out = pool(conv(inp))
print("Answer:", tuple(out.shape[2:]))

# q4
print("\nq4. Linear(128 -> 64) total params?")
linear = nn.Linear(128, 64)
print("Answer:", sum(p.numel() for p in linear.parameters()))
```

```python
# q5
print("\nq5. 1050 samples, batch=32, drop_last=False -> batches?")
print("Answer:", math.ceil(1050 / 32))

# q6
print("\nq6. CrossEntropyLoss for logits [2.5,1.0,0.1], target=0?")
logits = torch.tensor([[2.5, 1.0, 0.1]])
target = torch.tensor([0])
loss = nn.CrossEntropyLoss()(logits, target)
print("Answer:", round(loss.item(), 3))
```
